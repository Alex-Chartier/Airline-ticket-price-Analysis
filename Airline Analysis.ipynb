## Univariate Analysis

1. What do coach ticket prices look like? What are the high and low values? What would be considered the average? Does $500 seem like a good price for a coach ticket?

````python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
import scipy.stats as stats

## Read in Data
flight = pd.read_csv("flight.csv")
print(flight.head())

## Task 1

````

Prices for coach tickets range from 45 to almost 600 dollars with an average price of 375 dollars.

2. Now visualize the coach ticket prices for flights that are 8 hours long. What are the high, low, and average prices for 8-hour-long flights? Does a $500 dollar ticket seem more reasonable than before?

````python
## Task 2
plt.hist(flight.coach_price[flight.hours == 8])
plt.show()
plt.close()
print(np.mean(flight.coach_price[flight.hours == 8]))
print(np.median(flight.coach_price[flight.hours == 8]))
````

For 8 hour flights you start to see a very common price range of 300 to 500 dollars with nearly the same frequency in all bins

3. How are flight delay times distributed? Let's say there is a short amount of time between two connecting flights, and a flight delay would put the client at risk of missing their connecting flight. You want to better understand how often there are large delays so you can correctly set up connecting flights. What kinds of delays are typical?

````python
## Task 3
plt.hist(flight.delay, bins= 400)
plt.axis([0, 50, 0, 150000])
plt.show()
plt.close()
print(flight.delay.mean())
print(np.median(flight.delay))
print(max(flight.delay))
````

What this shows is that most flights will face a 10 minute delay however there is the possibility of the delay being almost a full day.

## Bivariate Analysis

4. Create a visualization that shows the relationship between coach and first-class prices. What is the relationship between these two prices? Do flights with higher coach prices always have higher first-class prices as well?

````python
## Task 4
plt.scatter(flight.coach_price, flight.firstclass_price)
plt.show()
plt.close()
print(stats.pearsonr(flight.coach_price, flight.firstclass_price))
````

This shows a very strong correlation between the price of coach tickets and the price of first class tickets with a pearson correlation of nearly 0.76.

5. What is the relationship between coach prices and inflight features &mdash; inflight meal, inflight entertainment, and inflight WiFi? Which features are associated with the highest increase in price?

````python
## Task 5
sns.histplot(flight, x = "coach_price", hue = flight.inflight_meal)
plt.show()
plt.close()


sns.histplot(flight, x = "coach_price", hue = flight.inflight_entertainment)
plt.show()
plt.close()


sns.histplot(flight, x = "coach_price", hue = flight.inflight_wifi)
plt.show()
plt.close()


````

You do see a rise in frequency of higher priced flights having inflight meals, entertainment, and wifi however this is still lacking a lot of depth considering it does not take into account the length of the flights which was already shown to have a high correlation with the price of tickets. With that caviat out of the way we can consider wifi and entertainment to be likely to increase price or to only be made available on longer flights since they have a larger skew.

6. How does the number of passengers change in relation to the length of flights?

````python
## Task 6
plt.scatter(flight.miles, flight.passengers)
plt.show()
plt.close()
````

From what can be seen with visualizations there is not much of any correlation between the number of passengers and the length of the flight. The most insight offered was that there is a slightly higher frequency of smaller passenger counts on the shortest of flights

## Multivariate Analysis

7. Visualize the relationship between coach and first-class prices on weekends compared to weekdays.

````python
## Task 7
sns.lmplot(x = 'coach_price', y = 'firstclass_price', hue = 'weekend', data = flight)
plt.show()
plt.close()
````

Here we see a near matching correlation continuing the trend that coach and first class ticket prices are following a similar pattern for their pricing scheme

8. How do coach prices differ for redeyes and non-redeyes on each day of the week?

````python
## Task 8
sns.boxplot(x = "day_of_week", y = "coach_price", hue = "redeye", data = flight)
plt.show()
plt.clf()
````

This suggest that the most expensive redeye flight cost as much or less as the least expensive non-redeye flight. As well it displays that sunday, friday, and saturday flights cost more on average
